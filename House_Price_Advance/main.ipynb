```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np

def plot_distributions(df, columns=None, cols_per_row=3, save_prefix='distribution'):
    """
    Plots the distribution (histogram + KDE) for numerical columns in a grid layout.
    Saves the cohesive plot grid directly to a PNG file.
    """
    # Automatically select all numerical columns if none are provided
    if columns is None:
        columns = df.select_dtypes(include=['number']).columns.tolist()
    
    num_plots = len(columns)
    if num_plots == 0:
        print("No numerical columns found to plot.")
        return
    
    # Calculate rows required based on the requested columns per row
    rows = math.ceil(num_plots / cols_per_row)
    
    # Create grid using subplots (avoids plt.figure() overlapping limitations)
    fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 4, rows * 3.5))
    
    # Uniformly handle single-plot vs multi-plot axes layouts
    if num_plots == 1:
        axes = [axes]
    else:
        axes = axes.flatten()
        
    for i, col in enumerate(columns):
        sns.histplot(data=df, x=col, kde=True, ax=axes[i], color='skyblue')
        axes[i].set_title(f'Distribution of {col}', fontsize=11)
        axes[i].set_xlabel('')
        axes[i].set_ylabel('Count / Density')
        
    # Automatically clean up and delete unused blank subplots in the grid boundary
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    output_filename = f"{save_prefix}_grid.png"
    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
    print(f"Distribution plot successfully saved as: {output_filename}")

def get_correlation(df, feature1, feature2, method='pearson'):
    """
    Calculates and returns the specific correlation coefficient between two columns.
    Supported methods: 'pearson', 'spearman', 'kendall'
    """
    if feature1 not in df.columns or feature2 not in df.columns:
        raise ValueError(f"Error: One or both features ('{feature1}', '{feature2}') do not exist in the DataFrame.")
        
    # Drop records containing missing values in either targeted column
    clean_df = df[[feature1, feature2]].dropna()
    
    if clean_df.empty:
        print("Warning: No overlapping non-null data exists between these two features.")
        return None
        
    # Calculate the correlation coefficient
    corr_value = clean_df[feature1].corr(clean_df[feature2], method=method)
    return corr_value
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_matrix(df, method='pearson', filename='correlation_matrix.png'):
    """
    Isolates all numerical features, calculates the correlation matrix, 
    and generates a half-triangle heatmap displaying ONLY strong correlations 
    where r > 0.7 or r < -0.7.
    """
    numeric_df = df.select_dtypes(include=['number'])
    
    if numeric_df.empty:
        print("No numerical columns found in the dataset to build a correlation matrix.")
        return
        
    corr_matrix = numeric_df.corr(method=method)
    num_features = len(corr_matrix.columns)
    
    # Scale figure footprint dynamically based on the volume of attributes
    grid_size = max(6, num_features * 0.8)
    fig, ax = plt.subplots(figsize=(grid_size, grid_size * 0.8))
    
    # --- CRITICAL MODIFICATION: Construct a Compound Mask ---
    # 1. np.triu strips out the redundant upper triangle mirror-image clutter.
    # 2. np.abs(corr_matrix) <= 0.7 flags weak relationships to be hidden (True).
    # 3. The '|' operator combines them so only strong correlations remain visible.
    upper_triangle_mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    weak_correlation_mask = np.abs(corr_matrix) <= 0.7
    combined_mask = upper_triangle_mask | weak_correlation_mask
    
    # Dynamically scale text size inside squares to prevent overlapping text
    font_size = max(7, 12 - num_features // 3)
    
    sns.heatmap(
        corr_matrix, 
        mask=combined_mask,   # Passes the filtered strong-relationship mask
        annot=True, 
        fmt=".2f", 
        cmap='coolwarm', 
        vmin=-1,       # Anchored lower boundary for negative correlations
        vmax=1,        # Anchored upper boundary for positive correlations
        center=0,      # Perfectly balanced mid-point color neutrality
        square=True, 
        linewidths=.5, 
        cbar_kws={"shrink": .8},
        ax=ax,
        annot_kws={"size": font_size}
    )
    
    ax.set_title(f"Strong Correlation Dashboard ($|r| > 0.7$)\n({method.capitalize()}'s $r$)", fontsize=14, pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Strong correlation matrix heatmap successfully saved as: {filename}")
```
