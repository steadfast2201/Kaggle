```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np

def plot_distributions(df, columns=None, cols_per_row=3, save_prefix='distribution'):
    """
    Plots the distribution (histogram + KDE) for numerical columns in a grid layout.
    Saves the cohesive plot grid directly to a PNG file.
    """
    # Automatically select all numerical columns if none are provided
    if columns is None:
        columns = df.select_dtypes(include=['number']).columns.tolist()
    
    num_plots = len(columns)
    if num_plots == 0:
        print("No numerical columns found to plot.")
        return
    
    # Calculate rows required based on the requested columns per row
    rows = math.ceil(num_plots / cols_per_row)
    
    # Create grid using subplots (avoids plt.figure() overlapping limitations)
    fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 4, rows * 3.5))
    
    # Uniformly handle single-plot vs multi-plot axes layouts
    if num_plots == 1:
        axes = [axes]
    else:
        axes = axes.flatten()
        
    for i, col in enumerate(columns):
        sns.histplot(data=df, x=col, kde=True, ax=axes[i], color='skyblue')
        axes[i].set_title(f'Distribution of {col}', fontsize=11)
        axes[i].set_xlabel('')
        axes[i].set_ylabel('Count / Density')
        
    # Automatically clean up and delete unused blank subplots in the grid boundary
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    output_filename = f"{save_prefix}_grid.png"
    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
    print(f"Distribution plot successfully saved as: {output_filename}")

def get_correlation(df, feature1, feature2, method='pearson'):
    """
    Calculates and returns the specific correlation coefficient between two columns.
    Supported methods: 'pearson', 'spearman', 'kendall'
    """
    if feature1 not in df.columns or feature2 not in df.columns:
        raise ValueError(f"Error: One or both features ('{feature1}', '{feature2}') do not exist in the DataFrame.")
        
    # Drop records containing missing values in either targeted column
    clean_df = df[[feature1, feature2]].dropna()
    
    if clean_df.empty:
        print("Warning: No overlapping non-null data exists between these two features.")
        return None
        
    # Calculate the correlation coefficient
    corr_value = clean_df[feature1].corr(clean_df[feature2], method=method)
    return corr_value
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_matrix(df, method='pearson', filename='correlation_matrix.png'):
    """
    Isolates all numerical features, calculates the correlation matrix, 
    and generates a half-triangle heatmap displaying ONLY strong correlations 
    where r > 0.7 or r < -0.7.
    """
    numeric_df = df.select_dtypes(include=['number'])
    
    if numeric_df.empty:
        print("No numerical columns found in the dataset to build a correlation matrix.")
        return
        
    corr_matrix = numeric_df.corr(method=method)
    num_features = len(corr_matrix.columns)
    
    # Scale figure footprint dynamically based on the volume of attributes
    grid_size = max(6, num_features * 0.8)
    fig, ax = plt.subplots(figsize=(grid_size, grid_size * 0.8))
    
    # --- CRITICAL MODIFICATION: Construct a Compound Mask ---
    # 1. np.triu strips out the redundant upper triangle mirror-image clutter.
    # 2. np.abs(corr_matrix) <= 0.7 flags weak relationships to be hidden (True).
    # 3. The '|' operator combines them so only strong correlations remain visible.
    upper_triangle_mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    weak_correlation_mask = np.abs(corr_matrix) <= 0.7
    combined_mask = upper_triangle_mask | weak_correlation_mask
    
    # Dynamically scale text size inside squares to prevent overlapping text
    font_size = max(7, 12 - num_features // 3)
    
    sns.heatmap(
        corr_matrix, 
        mask=combined_mask,   # Passes the filtered strong-relationship mask
        annot=True, 
        fmt=".2f", 
        cmap='coolwarm', 
        vmin=-1,       # Anchored lower boundary for negative correlations
        vmax=1,        # Anchored upper boundary for positive correlations
        center=0,      # Perfectly balanced mid-point color neutrality
        square=True, 
        linewidths=.5, 
        cbar_kws={"shrink": .8},
        ax=ax,
        annot_kws={"size": font_size}
    )
    
    ax.set_title(f"Strong Correlation Dashboard ($|r| > 0.7$)\n({method.capitalize()}'s $r$)", fontsize=14, pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Strong correlation matrix heatmap successfully saved as: {filename}")
```

## Train and Test XGBR

```python
X = df_train.drop(columns=["SalePrice"])
Y = df_train["SalePrice"]
```

```python
feature_names = X.columns
```

```python
print(feature_names)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
import numpy as np

# ----------------------------
# Feature Scaling (optional for XGBoost)
# ----------------------------
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ----------------------------
# Log-transform target
# ----------------------------
Y_train_log = np.log1p(Y_train)
Y_test_log = np.log1p(Y_test)

# ----------------------------
# Model
# ----------------------------
xgbr_model = XGBRegressor(
    n_estimators=150,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    objective="reg:squarederror"
)

# ----------------------------
# Train
# ----------------------------
xgbr_model.fit(X_train_scaled, Y_train_log)

# ----------------------------
# Predictions in log-space
# ----------------------------
train_pred_log = xgbr_model.predict(X_train_scaled)
test_pred_log = xgbr_model.predict(X_test_scaled)

# ----------------------------
# Competition Metric (Kaggle)
# ----------------------------
train_log_rmse = np.sqrt(
    mean_squared_error(Y_train_log, train_pred_log)
)

test_log_rmse = np.sqrt(
    mean_squared_error(Y_test_log, test_pred_log)
)

# ----------------------------
# Baseline (predict mean log price)
# ----------------------------
baseline_train_log = np.full(
    len(Y_train_log),
    Y_train_log.mean()
)

baseline_test_log = np.full(
    len(Y_test_log),
    Y_train_log.mean()
)

expected_train_log_rmse = np.sqrt(
    mean_squared_error(Y_train_log, baseline_train_log)
)

expected_test_log_rmse = np.sqrt(
    mean_squared_error(Y_test_log, baseline_test_log)
)

# ----------------------------
# Convert back to original scale
# ----------------------------
y_train_pred = np.expm1(train_pred_log)
y_test_pred = np.expm1(test_pred_log)

# ----------------------------
# R² on original SalePrice
# ----------------------------
train_r2 = r2_score(Y_train, y_train_pred)
test_r2 = r2_score(Y_test, y_test_pred)

# ----------------------------
# Results
# ----------------------------
print("=" * 60)
print("TRAIN METRICS")
print(f"Model Log-RMSE      : {train_log_rmse:.5f}")
print(f"Baseline Log-RMSE   : {expected_train_log_rmse:.5f}")
print(f"R² Score            : {train_r2:.5f}")

print("\n" + "=" * 60)
print("TEST METRICS")
print(f"Model Log-RMSE      : {test_log_rmse:.5f}")
print(f"Baseline Log-RMSE   : {expected_test_log_rmse:.5f}")
print(f"R² Score            : {test_r2:.5f}")
print("=" * 60)
```

```python
import optuna
import numpy as np
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

best_model = None

def objective(trial):

    global best_model

    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 1000),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.2, log=True),
        "max_depth": trial.suggest_int("max_depth", 2, 8),
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
        "gamma": trial.suggest_float("gamma", 0, 5),
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-5, 10, log=True),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-5, 10, log=True),
        "random_state": 42,
        "objective": "reg:squarederror",
    }

    model = XGBRegressor(**params)

    model.fit(
        X_train_scaled,
        Y_train_log,
        verbose=False
    )

    train_pred_log = model.predict(X_train_scaled)
    test_pred_log = model.predict(X_test_scaled)

    train_rmse = np.sqrt(
        mean_squared_error(Y_train_log, train_pred_log)
    )

    test_rmse = np.sqrt(
        mean_squared_error(Y_test_log, test_pred_log)
    )

    rmse_gap = abs(train_rmse - test_rmse)

    # Penalize overfitting
    score = test_rmse + 0.5 * rmse_gap

    trial.set_user_attr("train_rmse", train_rmse)
    trial.set_user_attr("test_rmse", test_rmse)
    trial.set_user_attr("rmse_gap", rmse_gap)

    return score


study = optuna.create_study(direction="minimize")

study.optimize(
    objective,
    n_trials=100,
    show_progress_bar=True
)

print("\nBest Trial")
print("-" * 50)

best_trial = study.best_trial

print("Combined Score :", best_trial.value)
print("Train Log-RMSE :", best_trial.user_attrs["train_rmse"])
print("Test Log-RMSE  :", best_trial.user_attrs["test_rmse"])
print("RMSE Gap       :", best_trial.user_attrs["rmse_gap"])

print("\nBest Parameters")
for k, v in best_trial.params.items():
    print(f"{k}: {v}")
```
